```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('dftrain5.csv')
```

```python
df.info()
```

```python
#Cleaning up DF
df.batter_stance = df.batter_stance.replace(['R', 'L'], [0,1])
df.top = df.top.replace([True, False], [1,0])
df.pitcher_throws=df.pitcher_throws.replace(['R', 'L'], [0,1])
```

```python
df.head()
```

```python

df= df.drop(columns= ['weather', 'year', 'Unnamed: 0', 'Unnamed: 0.1'])

```

```python
df.head()
```

```python
#getting how many pitches for each pitcher
df['pitcher_id'].value_counts()
```

```python
#keeping top 50 pitchers in terms of pitches thrown
vdf = pd.DataFrame(df['pitcher_id'].value_counts())
vdf.describe()
vdf.head(50)
```

```python
vdf['check'] = vdf['pitcher_id'] > 10161
```

```python
over_10161 = list(vdf.loc[vdf['check'], :].reset_index()['index'])
```

```python
df['check'] = df['pitcher_id'].isin(over_10161)
new_df = df.loc[df['check'], :]
new_df = new_df.drop(columns = ['check',])
new_df.info()
```

```python
df=new_df
```
